#Proper Visualization of the data set


```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=Superstore)
plt.show()
```

#Total Counts by State

```python
print(Superstore['State'].value_counts())
plt.figure(figsize=(15,8))
sns.countplot(x=Superstore['State'])
plt.xticks(rotation=90)
plt.show()
```

#Total Counts by Sub-Category 

```python
print(Superstore['Sub-Category'].value_counts())
plt.figure(figsize=(12,6))
sns.countplot(x=Superstore['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

#Checking Correlation

```python
fig,axes = plt.subplots(1,1,figsize=(9,6))
sns.heatmap(Superstore.corr(), annot= True)
plt.show()
```

#Sub-category Vs Profit

```python
plt.figure(figsize=(40,25))
sns.barplot(x=Superstore['Sub-Category'], y=Superstore['Profit'])
```

```python
sns.pairplot(Superstore)
```
